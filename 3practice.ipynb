```python
#7
# Функции для выполнения арифметических операций
def summ(x, y):
    return x + y

def sub(x, y):
    return x - y

def multy(x, y):
    return x * y

def div(x, y):
    if y == 0:  # Проверка деления на ноль
        return "Ошибка: Деление на ноль невозможно!"
    else:
        return x / y

# Запрашиваем у пользователя знак арифметического действия и два целых числа
operation = input("Введите знак арифметического действия (+, -, *, /): ")
x = int(input("Введите первое число (x): "))
y = int(input("Введите второе число (y): "))

# Выбираем и выполняем нужную операцию в зависимости от введенного знака
if operation == '+':
    result = summ(x, y)
    print(f'{x} + {y} = {result}')
elif operation == '-':
    result = sub(x, y)
    print(f'{x} - {y} = {result}')
elif operation == '*':
    result = multy(x, y)
    print(f'{x} * {y} = {result}')
elif operation == '/':
    result = div(x, y)
    print(f'{x} / {y} = {result}')
else:
    print("Ошибка: Неверный знак арифметического действия!")
```

```python
#10
import math

# Функция для расчета площади круга по радиусу
def circle_area(radius):
    return math.pi * radius ** 2

# Функция для расчета площади прямоугольника по длине и ширине
def rectangle_area(length, width):
    return length * width

# Функция для расчета площади треугольника по трем сторонам (формула Герона)
def triangle_area_heron(a, b, c):
    s = (a + b + c) / 2  # Полупериметр
    return math.sqrt(s * (s - a) * (s - b) * (s - c))

# Функция для расчета площади треугольника через основание и высоту
def triangle_area_base_height(base, height):
    return 0.5 * base * height

# Запрашиваем у пользователя тип фигуры
figure_type = input("Введите тип фигуры (круг, прямоугольник, треугольник): ").strip().lower()

# В зависимости от выбранного типа фигуры, запрашиваем соответствующие размеры и вычисляем площадь
if figure_type == "круг":
    radius = float(input("Введите радиус круга: "))
    area = circle_area(radius)
    print(f"Площадь круга с радиусом {radius} равна: {area:.2f}")
    
elif figure_type == "прямоугольник":
    length = float(input("Введите длину прямоугольника: "))
    width = float(input("Введите ширину прямоугольника: "))
    area = rectangle_area(length, width)
    print(f"Площадь прямоугольника размером {length} x {width} равна: {area:.2f}")

elif figure_type == "треугольник":
    method = input("Как вы хотите вычислить площадь треугольника? (по сторонам или по основанию и высоте): ").strip().lower()
    if method == "по сторонам":
        a = float(input("Введите длину первой стороны: "))
        b = float(input("Введите длину второй стороны: "))
        c = float(input("Введите длину третьей стороны: "))
        area = triangle_area_heron(a, b, c)
        print(f"Площадь треугольника со сторонами {a}, {b}, {c} равна: {area:.2f}")
    elif method == "по основанию и высоте":
        base = float(input("Введите длину основания треугольника: "))
        height = float(input("Введите высоту треугольника: "))
        area = triangle_area_base_height(base, height)
        print(f"Площадь треугольника с основанием {base} и высотой {height} равна: {area:.2f}")
    else:
        print("Ошибка: неверный метод вычисления площади треугольника!")
else:
    print("Ошибка: неверный тип фигуры!")
```

```python
#11
import math

# Лямбда-выражения для расчета площадей

# Площадь круга
circle_area = lambda r: math.pi * r**2

# Площадь прямоугольника
rectangle_area = lambda a, b: a * b

# Площадь треугольника по формуле Герона
triangle_area_heron = lambda a, b, c: math.sqrt((a + b + c) / 2 * ((a + b + c) / 2 - a) * ((a + b + c) / 2 - b) * ((a + b + c) / 2 - c))

# Площадь треугольника через основание и высоту
triangle_area_base_height = lambda base, height: 0.5 * base * height

def main():
    print("Выберите фигуру для расчета площади:")
    print("1. Круг")
    print("2. Прямоугольник")
    print("3. Треугольник")
    
    choice = input("Введите номер фигуры: ")

    if choice == '1':
        r = float(input("Введите радиус круга: "))
        print(f"Площадь круга: {circle_area(r):.2f}")
    
    elif choice == '2':
        a = float(input("Введите длину первой стороны прямоугольника: "))
        b = float(input("Введите длину второй стороны прямоугольника: "))
        print(f"Площадь прямоугольника: {rectangle_area(a, b):.2f}")

    elif choice == '3':
        print("Выберите метод расчета площади треугольника:")
        print("1. По трем сторонам")
        print("2. Через основание и высоту")
        
        triangle_choice = input("Введите номер метода: ")

        if triangle_choice == '1':
            a = float(input("Введите длину первой стороны: "))
            b = float(input("Введите длину второй стороны: "))
            c = float(input("Введите длину третьей стороны: "))
            print(f"Площадь треугольника (по формуле Герона): {triangle_area_heron(a, b, c):.2f}")
        
        elif triangle_choice == '2':
            base = float(input("Введите длину основания: "))
            height = float(input("Введите высоту: "))
            print(f"Площадь треугольника (по основанию и высоте): {triangle_area_base_height(base, height):.2f}")
        
        else:
            print("Неверный выбор метода для треугольника.")
    
    else:
        print("Неверный выбор фигуры.")

if __name__ == "__main__":
    main()
```

```python
#12
def find_divisors():
    # Запрашиваем у пользователя целое число
    num = int(input("Введите целое число: "))
    
    # Перебираем возможные делители от 2 до 10
    for i in range(2, 11):
        if num % i == 0:  # Проверяем, является ли i делителем числа num
            print(f"Делится на: {i}")

# Вызов функции
find_divisors()
```

```python
#13
def get_imt(m, h):
    # Перевод роста из см в метры
    h_meters = h / 100
    
    # Рассчет ИМТ по формуле
    imt = m / (h_meters ** 2)
    
    # Выводим рассчитанный ИМТ
    print(f"Ваш ИМТ: {imt:.2f}")
    
    # Определяем категорию по ИМТ
    if imt > 25:
        print("Избыточный вес")
    elif imt < 18.5:
        print("Недостаточный вес")
    else:
        print("Норма")

# Тесты
m = float(input("Введите массу тела (кг): "))
h = float(input("Введите рост (см): "))

# Вызов функции
get_imt(m, h)

# Входные данные:
# 100
# 190
# Выходные данные:
# Ваш ИМТ: 27.70
# Избыточный вес

# Входные данные:
# 80.2
# 184.3
# Выходные данные:
# Ваш ИМТ: 23.63
# Норма

# Входные данные:
# 55.7
# 175.9
# Выходные данные:
# Ваш ИМТ: 18.01
# Недостаточный вес
```

```python
#14
def calculate_final_grade(midterm, final_exam):
    # Рассчитываем итоговую оценку
    final_grade = 0.4 * midterm + 0.6 * final_exam
    
    # Проверяем, пройден ли курс
    if final_grade > 60:
        return "Зачет"
    else:
        return "Незачет"

# Тесты
midterm = float(input("Введите оценку за промежуточный экзамен: "))
final_exam = float(input("Введите оценку за итоговый экзамен: "))

# Вывод результата
result = calculate_final_grade(midterm, final_exam)
print(result)

# Входные данные:
# 50
# 70
# Выходные данные:
# Зачет

# Входные данные:
# 45
# 50
# Выходные данные:
# Незачет
```

```python
#15
def calculate_bonus(salary, performance):
    # Рассчитываем премию на основе производительности
    if performance > 90:
        return salary * 0.10  # 10% премии
    elif 80 <= performance <= 90:
        return salary * 0.05  # 5% премии
    else:
        return 0.0  # Нет премии

# Ввод данных от пользователя
salary = float(input("Введите заработную плату: "))
performance = int(input("Введите эффективность (в процентах): "))

# Вызываем функцию и выводим результат
bonus = calculate_bonus(salary, performance)
print(f"Премия: {bonus:.2f}")

# Тесты
# Входные данные: 1000.0, 95  Выходные данные: 100.0
# Входные данные: 2000.0, 85  Выходные данные: 100.0
# Входные данные: 1500.0, 75  Выходные данные: 0.0
# Входные данные: 1500.0, 80  Выходные данные: 75.0
# Входные данные: 500.0, 60   Выходные данные: 0.0
```

```python
#16
def month(month_number, language):
    # Словари с названиями месяцев на русском и английском языках
    months_ru = {
        1: "Январь", 2: "Февраль", 3: "Март", 4: "Апрель",
        5: "Май", 6: "Июнь", 7: "Июль", 8: "Август",
        9: "Сентябрь", 10: "Октябрь", 11: "Ноябрь", 12: "Декабрь"
    }
    
    months_en = {
        1: "January", 2: "February", 3: "March", 4: "April",
        5: "May", 6: "June", 7: "July", 8: "August",
        9: "September", 10: "October", 11: "November", 12: "December"
    }
    
    # Возвращаем название месяца в зависимости от языка
    if language == "ru":
        return months_ru.get(month_number, "Неверный номер месяца")
    elif language == "en":
        return months_en.get(month_number, "Invalid month number")
    else:
        return "Неверный язык"

print(month(int(input("Месяц / month")), input("Язык / language")))
```
